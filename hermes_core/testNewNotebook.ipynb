```python
import numpy as np
from hermes_core import Workspace
```

```python
# 1. Create dummy data
dummy_matrix = np.ones((100, 100, 100), dtype=np.uint16)

# 2. Instantiate your Workspace
ws = Workspace(matrix=dummy_matrix, voxel_size=1.0, name="TestBlock")
```

```python
# 3. Process it
ws.pad()
ws.generate_mesh()
mesh = ws.get_trimesh()

print(f"Successfully created a mesh for {ws.name}!")
print(f"Surface Area: {mesh.area}")
```

```python
# Cell 4
# 1. Extract a 50x50x50 subvolume from the corner (this will catch a quarter of the sphere)
print("Extracting subvolume...")
sub_ws = ws.extract_subvolume(corner=(25, 25, 25), dimensions=(50, 50, 50), sub_id=1)
print(f"Subvolume created: {sub_ws.name}")
print(f"Origin coordinates: {sub_ws.origin}, Shape: {sub_ws.matrix.shape}")

# 2. Pad the subvolume
sub_ws.pad(padding_size=1)
print(f"Shape after padding: {sub_ws.matrix.shape}")


if mesh:
    print("\n--- Mesh Properties ---")
    print(f"Is watertight? {mesh.is_watertight}")
    print(f"Volume: {mesh.volume:.2f}")
    print(f"Surface Area: {mesh.area:.2f}")
    print(f"Vertices: {len(mesh.vertices)}, Faces: {len(mesh.faces)}")
```
